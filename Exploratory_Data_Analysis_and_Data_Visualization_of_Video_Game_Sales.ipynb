# **About Dataset -- Dataset From KAGGLE**

# **Video Game Sales**

Analyze sales data from more than 16,500 games.

The script to scrape the data is available at https://github.com/GregorUT/vgchartzScrape.

It is based on BeautifulSoup using Python.

There are 16,598 records. 2 records were dropped due to incomplete information.


This dataset contains a list of video games with sales greater than 100,000 copies. It was generated by a scrape of vgchartz.com.

Fields include

Rank - Ranking of overall sales

Name - The games name

Platform - Platform of the games release (i.e. PC,PS4, etc.)

Year - Year of the game's release

Genre - Genre of the game

Publisher - Publisher of the game

NA_Sales - Sales in North America (in millions)

EU_Sales - Sales in Europe (in millions)

JP_Sales - Sales in Japan (in millions)

Other_Sales - Sales in the rest of the world (in millions)

Global_Sales - Total worldwide sales.

Reference -- https://www.kaggle.com/datasets/gregorut/videogamesales

# **Exploratory Data Analysis**


# ***1. Bar Chart***

**Global sales visualization against Platform and Genre.**

```python
df_platform = df.groupby('Platform', as_index=False)['Global_Sales'].sum()
df_platform
```

```python
df_platform.nlargest(31,'Global_Sales')
```

```python
plt.figure(figsize=(12,10))
sns.barplot(data=df_platform, x='Global_Sales', y='Platform')
plt.title('Total Global Sales In Each Platform', fontsize=18)
plt.xlabel ('Global Sales', fontsize=16)
plt.ylabel ('Platform', fontsize=16)
plt.show()
```

# From the visualization it can be concluded that,

**The platform named " PS2" had the highest sales globally.**

**After PS2, X360 had the second highest sales globally.**

**Platform DC had the lowest sales globally.**


```python
df_total = df.groupby('Genre', as_index=False)['Global_Sales'].sum()
df_total
```

```python
df_total.nlargest(12,'Global_Sales')
```

```python
plt.figure(figsize=(10,8))
sns.barplot(data=df_total, x='Global_Sales', y='Genre')
plt.title('Total Global Sales In Each Genre', fontsize=18)
plt.xlabel ('Global Sales', fontsize=16)
plt.ylabel ('Genre', fontsize=16)
plt.show()
```

# From the visualization it can be concluded that,

**Genre Action had the highest sales globally.**

**Genre Sports had the second highest sales globally and after it third is Shooter**.

**Genre Strategy had the lowest sales globally.** 

**Genre Misc and Platform have the kind of nearest sales globally**.

## *Visualization of Publisher column to see which company publishes the what number of games.For simplify taking publishers with game counts above 200 only.*

```python
plt.figure(figsize=(10,8))
publishers = df.Publisher.value_counts()
publishers = publishers[publishers.values > 200]
publishers.plot(kind = 'barh', color = 'Green')
plt.xlabel('Count')
plt.ylabel('Publishing Company')
plt.title('Publishing companies with more than 200 games')
plt.show()
```

# From the visualization it can be concluded that,

**Electronics Arts is the undisputed leader in terms of the video games sale count.**

**Activision, Namco Bandai Games, Ubisoft have almost similar sale count.**

**THQ, Nintendo, Sony Computer Entertainment have similar types of sale count.**

# ***2. Count plot***

**Maximum games sold using Countplot method on the basis of year**

```python
plt.figure(figsize=(30,10))
sns.countplot('Year', data = df, saturation = 0.70)
plt.title('Maximum Games sold on basis of Year')
plt.show()
```

# ***3. Pair Plot***

**Corellating among all Countries using Seaborn to perform pairplot and to plot the graph with matplotlib.**

```python
sns.pairplot(df.loc[0:,['NA_Sales','EU_Sales','JP_Sales', 'Other_Sales','Global_Sales']])
plt.show()
```

# Conclusion Upon performing the correlation among various countries :

**North America was the major market as the Global sales are highly correlated with it.**

**Europe was also an important region.**

**One intresting thing is Japanies sales were not correlated with any region's sales,We can assume that Japanies people have different taste, when it's about games.**

# ***4.Scatter Plot***





**Conclusion Upon performing the visualization on numeric sales of column for North America, Europe, Japan and Rest of the World**

```python
fig, axes = plt.subplots(2,2, figsize= (18, 14))
axes[0, 0].scatter(df.Year, df.NA_Sales, color = 'b')
axes[0, 0].set_title('North America Video Game Sales')
axes[0, 0].set_xlabel('Year')
axes[0, 0].set_ylabel('Sales in Million')

axes[0, 1].scatter(df.Year, df.EU_Sales, color = 'r')
axes[0, 1].set_title('Europe Video Game Sales')
axes[0, 1].set_xlabel('Year')
axes[0, 1].set_ylabel('Sales in Million')

axes[1, 0].scatter(df.Year, df.JP_Sales, color = 'g')
axes[1, 0].set_title('Japan Video Game Sales')
axes[1, 0].set_xlabel('Year')
axes[1, 0].set_ylabel('Sales in Million')

axes[1, 1].scatter(df.Year, df.Other_Sales, color = 'purple')
axes[1, 1].set_title('Other country Video Game Sales')
axes[1, 1].set_xlabel('Year')
axes[1, 1].set_ylabel('Sales in Million')
plt.show()
```

# From the above visualization we can conclude that, 

**The video game sales in the four different regions, the outliers in all the four regions are very few.**

**We can see that the sales have a higher number in the North American region**.

**The sales in other countries follow a distribution.**

# ***Heatmap***

***Heatmap is a good way to understand correlation***.

```python
cor = df.corr()
cor
```

```python
plt.figure(figsize=(12,10))
sns.heatmap(cor, annot= True,  cmap=plt.cm.CMRmap_r)
plt.show
```

# Here, firstly, for better understanding or simplization of the heatmap, the correlation between each column has been found out.

**1.From the visualization, it can be concluded that the correlation is 'Very Strong' between Global_Sales and NA_Sales, which means North America had the highest sales of video games compare to any other Countries.**

**2.The corrleation is 'Strong' among** 

**NA_Sales, EU_Sales and Other_Sales.**

**Global_Sales, JP_Sales and Other_Sales.**

**But JP_Sales has a strong corrleation only with Global_Sales.** 

**3.The corrleation is 'Moderate' among JP_Sales and**
**NA_Sales, EU_Sales.**

**4.The corrleation is 'Weak' between JP_Sales and Other_Sales.** 


**5. The corrleation is 'Negative/Negligible' Between Rank and Sales in all the four regions.**
